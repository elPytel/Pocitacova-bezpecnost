# Kdo jsem a jaká jsou práva superuživatele?

Spusťte následujíc buňku s příkazem `whoami` a zjistěte, pod jakým uživatelem jste aktuálně přihlášení.

````bash
whoami
````

Když spustíte příkaz `sudo whoami`, tak se buňka zasekne. Aby jste vykonávají zastavili zmáčkněte hranatý symbol "Stop" v levé horní části buňky.

Nevýhoda Jupyteru je, že neumí interaktivní příkazy, které čekají na vstup uživatele. 

Proto si otevřete terminál v okně VS Code. 

Horní lišta -> Terminál -> Nový terminál

A v něm napište příkaz:

```bash
sudo whoami
```

Budeme jej používat vždy když je potřeba něco vykonat s vyššími právy, například instalace programů, úprava systémových souborů nebo přístup k chráněným zdrojům.

*elevated privileges* znamená, že máte vyšší oprávnění než běžný uživatel, což vám umožňuje provádět akce, které by jinak byly omezené.

Další způsob jak obejít interaktivním vstup v Jupyteru je přesměrování vstupu z nějakého souboru, například:

Nasledující příkaz vytvoří soubor v tomto adresáři s názvem `.secret`. 

```bash
touch .secret
```

V Linuxu jsou přípony souborů nepovinné. O jaký soubor se jedná zjistí programy z hlavičky souboru, která obsahuje informace o typu souboru. 

Všechny soubory, které začínají tečkou jsou skryté. To znamená že v běžné adresářové struktuře je neuvidíte, pokud nezapnete zobrazení skrytých souborů. 

```bash
echo "tajný obsah" > .secret
```

````bash
touch .secret
````

````bash
touch grass
````

Pak můžeme načíst obsah tohoto souboru pomocí příkazu `cat` do proměnné `PASSWORD` a následně ji použít v příkazu `sudo`:

Jelikož jste v kurzu počítačové bezpečnosti, tak je vám jistě jasné, že ukládat hesla v plain-textu na svém počítači není bezpečné.

Nyní si to jednou vyzkoušejte. V budoucnu bude na vás, zda použijete pro spouštění příkazů terminál (a výsledek sem překopírujete), nebo Jupyter.

````bash
# nastení hesla pro sudo
PASSWORD=$(cat .secret)
````

`apt` je zkratka pro "Advanced Package Tool" a je to nástroj pro správu balíčků v Linuxu na distribucích odvozených od Debianu (např. Ubuntu). Umožňuje vám instalovat, aktualizovat a odstraňovat software z repozitářů.

Repozitář je místo, kde jsou uloženy balíčky software. Linux je má již velmi dlouho. S obdobnou funkcionalitou později přišel MacOS (AppStore) a nakonec i Windows (Microsoft Store).

Příkaz:

```bash
sudo apt-get update
```

Aktualizuje seznam dostupných balíčků a jejich verzí z repozitářů. Je důležité tento příkaz spustit před instalací nového software (bohatě stačí jednou za den), aby systém věděl o nejnovějších verzích.

````bash
echo $PASSWORD | sudo -S apt-get update
````

Nyní použijeme příkaz `rm` k odstranění souboru `.secret`, abychom zajistili, že tajný obsah nebude dále přístupný.

````bash
rm .secret
````

## Nápověda a manuály (RTFM)
Ale co všechny ty příkazy vůbec znamenají a jak zjistit k čemu slouží?

```bash
man
```

man je zkratka pro "manual" a slouží k zobrazení nápovědy a dokumentace k různým příkazům v Linuxu. Když spustíte `man rm`, zobrazí se vám podrobný popis příkazu `rm`, jeho možnosti a použití.


![](../../assets/thank-you-mansplaining.gif)

````bash
man rm
````

Další možnost je se zeptat samotného příkazu na nápovědu pomocí přepínače `--help`.

````bash
touch --help
````

````bash
echo $PASSWORD | sudo apt install -y sl cmatrix
````

Jaktože šel následují příkaz spustit i když jsme odstranili soubor `.secret`?

Používáme proměnnou `PASSWORD`, která již obsahuje načtené heslo. I když jsme odstranili soubor, proměnná stále drží tuto hodnotu, protože byla načtena do paměti před odstraněním souboru.

Zapomeneme obsah proměnné `PASSWORD`

````bash
unset PASSWORD
````

````bash
echo $PASSWORD | sudo whoami
````

Pořád to funguje?!

Linux je chytrý a ví že pokud jste nedávno zadali heslo správce, tak vám ještě chvíli dovolí spouštět příkazy s `sudo` bez zadání hesla. Toto je bezpečnostní opatření, které umožňuje administrátorům provádět více operací bez opakovaného zadávání hesla, ale zároveň to není trvalé a po určité době (obvykle 15 minut) budete muset znovu zadat heslo.

````bash
echo $PASSWORD
````

````bash
sudo whoami
````
